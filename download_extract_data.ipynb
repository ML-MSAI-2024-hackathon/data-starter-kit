```python
import zipfile
import os

# Path to the downloaded zip file
zip_path = '/gdrive/MyDrive/ML-MSAI-2024-hackathon-data-colab/modgen-data-anon.zip'
# Target directory to extract the contents
extract_path = '/gdrive/MyDrive/ML-MSAI-2024-hackathon-data-colab/'

# Ensure the target directory exists
os.makedirs(extract_path, exist_ok=True)

# Unzipping the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Unzipping completed successfully!")
```
